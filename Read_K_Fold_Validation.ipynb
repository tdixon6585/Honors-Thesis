```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
f = pd.read_csv('./K_Fold_Accuracies.txt', sep='\t', index_col=False)
```

```python
#for i, txt in enumerate(f['Layers']):
#    plt.annotate(txt, (f['learning_rate'].iloc[i], f['Btest_acc'].iloc[i]), 
#                 textcoords="offset points",
#                 xytext=(0,10),
#                 ha='center')

plt.plot(f['learning_rate'], f['Btest_acc'], 'o', label='Training')
plt.plot(f['learning_rate'], f['Btrain_acc'], 'o', label='Testing')
plt.xscale('log')
plt.ylim(0,1)
plt.title("Basic_NN Accuracies vs. Learning Rate")
plt.legend()
plt.show()

plt.plot(f['learning_rate'], f['Ktrain_acc'], 'o', label='Training')
plt.plot(f['learning_rate'], f['Ktest_acc'], 'o', label='Testing')
plt.xscale('log')
plt.ylim(0,1)
plt.title("Keras_NN Accuracies vs. Learning Rate")
plt.legend()
plt.show()
```

```python
for l in f['Layers'].unique():
    plt.plot(f.loc[f['Layers']==l, 'learning_rate'],f.loc[f['Layers']==l, 'Btest_acc'], 'o', label='Basic_Test')
    plt.plot(f.loc[f['Layers']==l, 'learning_rate'],f.loc[f['Layers']==l, 'Btrain_acc'], 'o', label='Basic_Train')
    plt.plot(f.loc[f['Layers']==l, 'learning_rate'],f.loc[f['Layers']==l, 'Ktest_acc'], 'o', label='Keras_Test')
    plt.plot(f.loc[f['Layers']==l, 'learning_rate'],f.loc[f['Layers']==l, 'Ktrain_acc'], 'o', label='Keras_Train')
    plt.xscale('log')
    plt.ylim(0,1)
    plt.title(f'{l} Accuracy vs. Learning Rate')
    plt.xlabel('Learning Rate')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()
```
